```python
from sklearn.utils import resample
```

```python
# create two differenct dataframe of majority and minority class
data_majority = data[(data['doripenem']==0)]
data_minority = data[(data['doripenem']==1)]
```

```python
# upsample minority class
data_minority_upsampled = resample(data_minority,
                                  replace=True,
                                  n_samples=75,
                                  random_state=42)
```

```python
data_upsampled = pd.concat([data_minority_upsampled, data_majority])
```
